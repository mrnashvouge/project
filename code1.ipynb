```python
import sys
import collections, numpy
import numpy as np
numpy.set_printoptions(threshold=sys.maxsize)
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import DBSCAN
from collections import Counter
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
x = [ ]
y =[ ]
count = 0
file = open('C:/Users/Vogue/Desktop/data4.txt','r')
for line in file :
    count += 1
    if count >= 2 :
        d= line.split(",")
        x.append(float(d[2]))
        x.append(float(d[3]))
        a = d[7].split("/")
        if a[1] == '10' :
            x.append(float(a[0])/100)
        if a[1] == '11' :
            b = float(a[0])+31
            x.append(float(b)/100)
        if a[1] == '12':
            b = float(a[0])+61
            x.append(float(b)/100)
        if a[1] == '1':
            b = float(a[0])+92
            x.append(float(b)/100)
        if a[1] == '2':
            b = float(a[0])+123
            x.append(float(b)/100)
        if a[1] == '3':
            b = float(a[0])+151
            x.append(float(b)/100)
        if a[1] == '4':
            b = float(a[0])+182
            x.append(float(b)/100)
        y.append(x)
        x = []
        if count == 200000:
            break
data = np.array(y)

# Determine parameter for DBSCAN
model = DBSCAN(eps=0.02, min_samples=2)
model.fit_predict(data)
pred = model.fit_predict(data)
labels = model.labels_
n_noise_ = list(labels).count(-1)

#result from DBSCAN
print("number of cluster found: {}".format(len(set(model.labels_))))
print('cluster for each point: ', model.labels_)
print('Estimated number of noise points: %d' % n_noise_)
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(data, labels))

#data graph 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:,0], data[:,1], data[:,2]*100, s=2)
ax.view_init(azim=135)
ax.set_xlabel('lat(X)')
ax.set_ylabel('lon(Y)')
ax.set_zlabel('day(Z)')
plt.show()

# DBSCAN clustering graph 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:,0], data[:,1], data[:,2]*100, c=model.labels_, s=2)
ax.view_init(azim=135)
ax.set_xlabel('lat(X)')
ax.set_ylabel('lon(Y)')
ax.set_zlabel('day(Z)')
plt.show()

#At parameter epsilon=0.02 minpoints= 2
print('points in each cluster:' ,Counter(labels))

count1=0
n=0
p = [ ]
k = [ ]
for line in labels :
    count1 += 1
    if count1 >= 1 :
        u=labels[n]
        t=data[n]
        t=np.append(t,u)
        p=np.append(p,t)
        n +=1
        k.append(p)
        p=[ ]
    if count1 == 200000:
        break

# group 6306 (group with the biggest amount of hotspots)
count2=0
n=0
e=[ ]
r=[ ]
n1=0
for line in k :
    count2 += 1
    if count2 >= 1 :
        l=k[n][3]
        if float(l) == 6306:
            e.append(k[n][0])
            e.append(k[n][1])
            e.append(k[n][2])
            p=[e]
    r=np.append(r,[e])
    e=[ ]
    n+=1
    if count2 == 200000:
        break
count3=0
x1= [ ]
y1=[ ]
n=0
for line in r :
    count3 += 1
    if count3 >= 1 :
        x1.append(r[n])
        x1.append(r[n+1])
        x1.append(r[n+2])
        y1.append(x1)
        x1=[ ]
    n+=3
    if count3>= 2155 :
        break
data = np.array(y1)
print(data)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:,0], data[:,1], data[:,2]*100, s=5)
ax.view_init(azim=135)
ax.set_xlabel('lat(X)')
ax.set_ylabel('lon(Y)')
ax.set_zlabel('day(Z)')
print('day: ',Counter(r))
#graph group 6306 (the biggest amount of hotspots)
plt.show()





```
